## MERGE TRAINING DATA

### Parameter & Konfigurasi Periode

```python
import numpy as np
import pandas as pd

# --- Periode observasi (hanya jam 06:00, 12:00, 18:00) ---
OBS_HOURS = [6, 12, 18]

TRAIN_START = pd.Timestamp("2023-01-01 06:00:00")
TRAIN_END   = pd.Timestamp("2025-09-18 18:00:00")
TEST_START  = pd.Timestamp("2025-09-19 06:00:00")
TEST_END    = pd.Timestamp("2026-05-18 18:00:00")

# --- Rekayasa fitur historis ---
N_LAGS = 11                 # menghasilkan wl_t, wl_t-1, ..., wl_t-11
ROLLING_WINDOWS = [4, 8, 28]  # dalam satuan jumlah observasi (bukan jam),
                              # ~1 hari, ~2-3 hari, ~9-10 hari (3 obs/hari)

# --- Ukuran window temporal untuk tensor X ---
TIME_WINDOW = 12  # jumlah langkah observasi historis (t-11 ... t) yang
                   # disusun sebagai sumbu waktu pada tensor X

# --- Graph ---
MAX_DOWNSTREAM_HOPS = 3000  # batas penelusuran segmen sungai NEXT_DOWN
                             # agar traversal tidak berjalan tak terbatas

print("Parameter siap.")
print("Train :", TRAIN_START, "-", TRAIN_END)
print("Test  :", TEST_START, "-", TEST_END)
```

### Urutan Node (Konsisten di Seluruh Timestep)

```python
# Urutan node HARUS konsisten di setiap timestep & antara X, edge_index, dan y.
# Gunakan urutan alfabetis dari nama_pos pada koordinat sebagai referensi utama.

node_order = sorted(coordinate["nama_pos"].unique().tolist())
num_nodes = len(node_order)
node_to_idx = {name: i for i, name in enumerate(node_order)}

# Validasi kelengkapan pos di setiap sumber data
pos_train = set(train_data["nama_pos"].unique())
pos_env = set(environment_data["nama_pos"].unique())
pos_coord = set(coordinate["nama_pos"].unique())

missing_in_train = pos_coord - pos_train
missing_in_env = pos_coord - pos_env

print(f"Jumlah node (pos): {num_nodes}")
if missing_in_train:
    print("Peringatan: pos tanpa data train:", missing_in_train)
if missing_in_env:
    print("Peringatan: pos tanpa data lingkungan:", missing_in_env)

coordinate_idx = coordinate.set_index("nama_pos").loc[node_order]
```

### Membangun Graph Konektivitas Sungai (HydroRIVERS)

Setiap pos disnap ke segmen sungai HydroRIVERS terdekat, lalu relasi hulu-hilir ditelusuri melalui atribut `NEXT_DOWN` untuk membentuk `edge_index` (dan `edge_weight` berupa panjang sungai antar pos, dalam km). Node yang tidak memiliki koneksi sungai langsung ke node lain (misal karena letaknya di anak sungai yang tidak tercakup rute NEXT_DOWN antar-pos) disambungkan memakai jarak spasial (haversine) sebagai fallback agar graph tetap terhubung. Graph dibuat dua arah (undirected) agar message passing GNN bisa memanfaatkan sinyal dari hulu maupun hilir.

```python
import geopandas as gpd
from shapely.geometry import Point

# --- 1) Snap setiap pos ke segmen sungai HydroRIVERS terdekat ---
coord_clean = coordinate_idx.reset_index().dropna(subset=["latitude", "longitude"])

stations_gdf = gpd.GeoDataFrame(
    coord_clean,
    geometry=gpd.points_from_xy(coord_clean["longitude"], coord_clean["latitude"]),
    crs="EPSG:4326",
)

if gdf.crs is None:
    gdf = gdf.set_crs("EPSG:4326")

gdf_m = gdf.to_crs("EPSG:3857")
stations_m = stations_gdf.to_crs("EPSG:3857")

# sertakan SEMUA kolom atribut HydroRIVERS (bukan cuma HYRIV_ID/NEXT_DOWN/LENGTH_KM)
# supaya nanti bisa dipakai sebagai node feature statis
hydro_cols_all = [c for c in gdf.columns if c != "geometry"]
gdf_m_sub = gdf_m[hydro_cols_all + ["geometry"]].reset_index(drop=True)
stations_m = stations_m.reset_index(drop=True)

snapped = gpd.sjoin_nearest(
    stations_m, gdf_m_sub,
    how="left", distance_col="dist_to_river_m",
)
snapped = snapped.drop(columns=["index_right"], errors="ignore")
snapped = snapped.drop_duplicates(subset="nama_pos", keep="first")

pos_to_hyriv = dict(zip(snapped["nama_pos"], snapped["HYRIV_ID"]))
hyriv_to_pos = {v: k for k, v in pos_to_hyriv.items()}

print("Contoh pemetaan pos -> segmen HydroRIVERS terdekat:")
print(snapped[["nama_pos", "HYRIV_ID", "dist_to_river_m"]].head())

# --- 2) Bangun peta konektivitas antar segmen sungai (hulu -> hilir) ---
next_down_map = dict(zip(gdf["HYRIV_ID"], gdf["NEXT_DOWN"]))
length_map = dict(zip(gdf["HYRIV_ID"], gdf["LENGTH_KM"]))


def trace_downstream_station(start_hyriv_id):
    """Telusuri rantai NEXT_DOWN dari suatu segmen sampai menemukan
    segmen lain yang merupakan lokasi sebuah pos (stasiun) hilir terdekat.
    Mengembalikan (nama_pos_hilir, total_panjang_km) atau (None, None)."""
    current = next_down_map.get(start_hyriv_id, None)
    total_length_km = 0.0
    hops = 0
    visited = set()
    while current is not None and current != 0 and hops < MAX_DOWNSTREAM_HOPS:
        if current in visited:
            break  # hindari infinite loop akibat data siklik
        visited.add(current)
        total_length_km += length_map.get(current, 0.0) or 0.0
        if current in hyriv_to_pos:
            return hyriv_to_pos[current], total_length_km
        current = next_down_map.get(current, None)
        hops += 1
    return None, None


edges = []          # list of (src_idx, dst_idx, weight_km, is_fallback)
connected_pairs = set()

for pos_a in node_order:
    hyriv_a = pos_to_hyriv.get(pos_a)
    if hyriv_a is None:
        continue
    pos_b, dist_km = trace_downstream_station(hyriv_a)
    if pos_b is not None and pos_b != pos_a:
        edges.append((node_to_idx[pos_a], node_to_idx[pos_b], dist_km, False))
        connected_pairs.add(frozenset((pos_a, pos_b)))

# --- 3) Fallback: sambungkan node yang belum punya edge sama sekali
#         (baik sebagai hulu maupun hilir) dengan tetangga spasial terdekat,
#         supaya graph tetap terhubung. Bobot memakai jarak haversine (km). ---
def haversine_km(lat1, lon1, lat2, lon2):
    r = 6371.0
    p1, p2 = np.radians(lat1), np.radians(lat2)
    dphi = np.radians(lat2 - lat1)
    dlmb = np.radians(lon2 - lon1)
    a = np.sin(dphi / 2) ** 2 + np.cos(p1) * np.cos(p2) * np.sin(dlmb / 2) ** 2
    return 2 * r * np.arcsin(np.sqrt(a))

connected_nodes = set()
for a, b, w, fb in edges:
    connected_nodes.add(a)
    connected_nodes.add(b)

isolated_nodes = [i for i in range(num_nodes) if i not in connected_nodes]
for i in isolated_nodes:
    pos_i = node_order[i]
    lat_i, lon_i = coordinate_idx.loc[pos_i, ["latitude", "longitude"]]
    dists = []
    for j, pos_j in enumerate(node_order):
        if j == i:
            continue
        lat_j, lon_j = coordinate_idx.loc[pos_j, ["latitude", "longitude"]]
        dists.append((haversine_km(lat_i, lon_i, lat_j, lon_j), j))
    dists.sort()
    nearest_j = dists[0][1]
    edges.append((i, nearest_j, dists[0][0], True))

# --- 4) Jadikan graph tak berarah (dua arah) agar message-passing GNN
#         bisa memakai informasi hulu<->hilir; bobot sama untuk kedua arah. ---
src_list, dst_list, w_list = [], [], []
for a, b, w, fb in edges:
    src_list += [a, b]
    dst_list += [b, a]
    w_list += [w, w]

edge_index = np.array([src_list, dst_list], dtype=np.int64)          # (2, num_edges)
edge_weight = np.array(w_list, dtype=np.float32)                      # (num_edges,)

print(f"Jumlah edge (searah hulu-hilir sebelum digandakan): {len(edges)}")
print(f"edge_index shape: {edge_index.shape}, edge_weight shape: {edge_weight.shape}")
print(f"Node tanpa pasangan sungai langsung (pakai fallback jarak): "
      f"{[node_order[i] for i in isolated_nodes]}")

# --- 5) Atribut HydroRIVERS sebagai node feature statis ---
hydrorivers_numeric_cols = [
    c for c in gdf.columns
    if c not in ("geometry", "HYRIV_ID", "NEXT_DOWN")
    and pd.api.types.is_numeric_dtype(gdf[c])
]
print("Atribut HydroRIVERS yang dipakai sebagai node feature:", hydrorivers_numeric_cols)

river_attrs = snapped.set_index("nama_pos")[hydrorivers_numeric_cols].reindex(node_order)
river_attrs.columns = [f"river_{c.lower()}" for c in river_attrs.columns]
```

### Filter ke Jam Observasi & Menangani Data Lingkungan yang Kosong

Data lingkungan (per jam) difilter hanya pada jam observasi (06:00, 12:00, 18:00). Banyak nilai kosong terutama di ujung periode data (mendekati 2026-05-18) ditangani dengan **forward-fill per pos** (hanya memakai data masa lalu, sehingga tidak terjadi data leakage). Sisa NaN di awal deret (sebelum observasi valid pertama suatu pos) diisi memakai rata-rata kolom tsb yang dihitung **hanya dari periode train**, supaya tidak bocor informasi dari periode test.

```python
train_data["datetime"] = pd.to_datetime(train_data["datetime"])
environment_data["datetime"] = pd.to_datetime(environment_data["datetime"])

obs_datetimes = pd.date_range(TRAIN_START, TEST_END, freq="6h")
obs_datetimes = obs_datetimes[obs_datetimes.hour.isin(OBS_HOURS)]

train_obs = train_data[train_data["datetime"].isin(obs_datetimes)].copy()
env_obs = environment_data[environment_data["datetime"].isin(obs_datetimes)].copy()

# kerangka lengkap (datetime x nama_pos) supaya celah data eksplisit sebagai NaN
full_index = pd.MultiIndex.from_product([obs_datetimes, node_order],
                                         names=["datetime", "nama_pos"])
env_full = (
    env_obs.set_index(["datetime", "nama_pos"])
    .reindex(full_index)
    .reset_index()
)

env_feature_cols = [
    "rainfall_mm", "humidity_pct", "wind_direction_deg", "dew_point_c",
    "cloud_cover_pct", "temperature_c", "wind_speed_kmh",
    "rainfall_openmeteo_mm", "rainfall_max_24h_mm", "solar_radiation_mj_m2",
    "soil_moisture_0_7cm", "soil_moisture_7_28cm", "soil_moisture_28_100cm",
    "soil_moisture_100_255cm", "surface_pressure_hpa", "pressure_msl_hpa",
    "built_surface_m2", "landcover_class",
    "rmm1", "rmm2", "mjo_phase", "mjo_amplitude", "mjo_active", "nino_34",
]
# landcover_name (kategorikal) tidak dipakai langsung karena sudah terwakili
# secara numerik oleh landcover_class; menghindari duplikasi informasi.

n_missing_before = env_full[env_feature_cols].isna().sum().sum()

# Isi kekosongan HANYA dengan forward-fill per pos (causal, tidak memakai
# data masa depan) -> aman dari data leakage. Ini juga menangani banyaknya
# data lingkungan yang kosong di ujung periode (2026-05-18).
env_full = env_full.sort_values(["nama_pos", "datetime"])
env_full[env_feature_cols] = (
    env_full.groupby("nama_pos")[env_feature_cols].ffill()
)

# Sisa NaN (hanya mungkin terjadi di AWAL deret suatu pos, sebelum ada
# observasi valid pertama) diisi dengan mean kolom tsb dari periode TRAIN
# saja, supaya tidak bocor informasi dari periode test.
train_mask_for_fill = env_full["datetime"] <= TRAIN_END
fallback_means = env_full.loc[train_mask_for_fill, env_feature_cols].mean()
env_full[env_feature_cols] = env_full[env_feature_cols].fillna(fallback_means)

n_missing_after = env_full[env_feature_cols].isna().sum().sum()
print(f"Nilai kosong pada fitur lingkungan: {n_missing_before} -> {n_missing_after} "
      "(setelah forward-fill per pos + fallback mean periode train)")
```

### Fitur Historis (Lag) & Rolling Statistics Ketinggian Air

Semua lag dan rolling statistics dihitung murni dari data hingga waktu `t` (tidak menggunakan data masa depan). Rolling statistics dihitung dari nilai yang sudah di-shift 1 langkah agar tidak menyertakan `wl_t` itu sendiri dalam window.

```python
wl_long = train_obs[["datetime", "nama_pos", "tma_mdpl"]].copy()
wl_long = (
    wl_long.set_index(["datetime", "nama_pos"])
    .reindex(full_index)
    .reset_index()
    .sort_values(["nama_pos", "datetime"])
)
wl_long = wl_long.rename(columns={"tma_mdpl": "wl_t"})

g = wl_long.groupby("nama_pos")["wl_t"]

# lag features: wl_t, wl_t-1, ..., wl_t-11 (semua berbasis data <= t, tidak leak)
for lag in range(1, N_LAGS + 1):
    wl_long[f"wl_t-{lag}"] = g.shift(lag)

# rolling statistics dihitung dari nilai <= t (shift(1) dulu supaya window
# tidak menyertakan wl_t itu sendiri, murni riwayat historis)
wl_shifted = g.shift(1)
wl_long["_wl_shift1"] = wl_shifted
gshift = wl_long.groupby("nama_pos")["_wl_shift1"]
for w in ROLLING_WINDOWS:
    wl_long[f"wl_roll{w}_mean"] = gshift.transform(lambda s: s.rolling(w, min_periods=w).mean())
    wl_long[f"wl_roll{w}_std"] = gshift.transform(lambda s: s.rolling(w, min_periods=w).std())
    wl_long[f"wl_roll{w}_min"] = gshift.transform(lambda s: s.rolling(w, min_periods=w).min())
    wl_long[f"wl_roll{w}_max"] = gshift.transform(lambda s: s.rolling(w, min_periods=w).max())
wl_long = wl_long.drop(columns=["_wl_shift1"])

lag_cols = [f"wl_t-{lag}" for lag in range(1, N_LAGS + 1)]
rolling_cols = [c for c in wl_long.columns if c.startswith("wl_roll")]

print(f"Fitur lag dibuat: {['wl_t'] + lag_cols}")
print(f"Fitur rolling dibuat: {rolling_cols}")
```

### Fitur Waktu Siklik & Fitur Spasial

```python
time_feat = pd.DataFrame({"datetime": obs_datetimes})
time_feat["hour_sin"] = np.sin(2 * np.pi * time_feat["datetime"].dt.hour / 24)
time_feat["hour_cos"] = np.cos(2 * np.pi * time_feat["datetime"].dt.hour / 24)
doy = time_feat["datetime"].dt.dayofyear
days_in_year = np.where(time_feat["datetime"].dt.is_leap_year, 366, 365)
time_feat["dayofyear_sin"] = np.sin(2 * np.pi * doy / days_in_year)
time_feat["dayofyear_cos"] = np.cos(2 * np.pi * doy / days_in_year)

spatial_feat = coordinate_idx[["latitude", "longitude"]].reset_index()
```

### Menggabungkan Seluruh Node Feature (Long Format)

Seluruh fitur (lingkungan, lag & rolling ketinggian air, waktu siklik, koordinat, dan atribut HydroRIVERS) digabung menjadi satu tabel panjang berbasis `(datetime, nama_pos)`. Target `y` didefinisikan sebagai ketinggian air pos tsb pada langkah observasi berikutnya (`t+1`).

```python
node_features = wl_long.merge(env_full, on=["datetime", "nama_pos"], how="left")
node_features = node_features.merge(time_feat, on="datetime", how="left")
node_features = node_features.merge(spatial_feat, on="nama_pos", how="left")
node_features = node_features.merge(
    river_attrs.reset_index(), on="nama_pos", how="left"
)

# target y = wl pada t+1 (langkah observasi berikutnya, per pos)
node_features = node_features.sort_values(["nama_pos", "datetime"])
node_features["y_next"] = node_features.groupby("nama_pos")["wl_t"].shift(-1)

feature_cols = (
    ["wl_t"] + lag_cols + rolling_cols
    + env_feature_cols
    + ["hour_sin", "hour_cos", "dayofyear_sin", "dayofyear_cos"]
    + ["latitude", "longitude"]
    + list(river_attrs.columns)
)
num_features = len(feature_cols)
print(f"Total node feature per timestep: {num_features}")
print(feature_cols)
```

### Menyusun Tensor Graph Time Series & Membuang Sample Tidak Lengkap

Setiap sample merepresentasikan seluruh jaringan sungai (30 pos) pada satu window waktu (`time_window` langkah observasi historis, termasuk `t`), dengan urutan node yang selalu mengikuti `node_order`. Sample dibuang apabila ada node yang kekurangan fitur, lag, atau target pada window tsb.

```python
# ======================================================================
# CELL (revisi): pivot menjadi tensor + split train (supervised) & 
#                 kerangka test (untuk inference autoregresif)
# ======================================================================
wl_related_cols = ["wl_t"] + lag_cols + rolling_cols
non_wl_cols = [c for c in feature_cols if c not in wl_related_cols]

# --- TRAIN: tetap wajib lengkap semua (fitur + lag + rolling + target) ---
train_period_mask = (node_features["datetime"] >= TRAIN_START) & (node_features["datetime"] <= TRAIN_END)
train_completeness = (
    node_features.loc[train_period_mask]
    .set_index(["datetime", "nama_pos"])[feature_cols + ["y_next"]]
    .notna().all(axis=1)
    .groupby(level=0).all()
)
train_valid_timesteps = pd.DatetimeIndex(sorted(train_completeness[train_completeness].index))
print(f"Timestep TRAIN valid (lengkap semua node): {len(train_valid_timesteps)}")

# --- TEST: hanya wajib fitur non-wl (lingkungan/waktu/spasial/HydroRIVERS) lengkap ---
test_period_mask = (node_features["datetime"] >= TEST_START) & (node_features["datetime"] <= TEST_END)
test_completeness = (
    node_features.loc[test_period_mask]
    .set_index(["datetime", "nama_pos"])[non_wl_cols]
    .notna().all(axis=1)
    .groupby(level=0).all()
)
test_valid_timesteps = pd.DatetimeIndex(sorted(test_completeness[test_completeness].index))
print(f"Timestep TEST valid (fitur non-wl lengkap semua node): {len(test_valid_timesteps)}")

# --- panel 3D untuk seluruh horizon (train+test) ---
panel = (
    node_features.set_index(["datetime", "nama_pos"])[feature_cols]
    .reindex(pd.MultiIndex.from_product([obs_datetimes, node_order], names=["datetime", "nama_pos"]))
)
panel_arr = panel.values.reshape(len(obs_datetimes), num_nodes, num_features).astype(np.float32)
datetime_to_pos = {dt: i for i, dt in enumerate(obs_datetimes)}
wl_col_positions = [feature_cols.index(c) for c in wl_related_cols]

target_series = (
    node_features.set_index(["datetime", "nama_pos"])["y_next"]
    .reindex(pd.MultiIndex.from_product([obs_datetimes, node_order], names=["datetime", "nama_pos"]))
)
target_arr = target_series.values.reshape(len(obs_datetimes), num_nodes).astype(np.float32)


def build_train_samples(target_datetimes):
    """Untuk TRAIN: semua fitur+target harus lengkap (supervised biasa)."""
    X_list, y_list, used_dt = [], [], []
    for dt in target_datetimes:
        t_pos = datetime_to_pos[dt]
        start = t_pos - (TIME_WINDOW - 1)
        if start < 0:
            continue
        window = panel_arr[start:t_pos + 1]
        y_t = target_arr[t_pos]
        if np.isnan(window).any() or np.isnan(y_t).any():
            continue
        X_list.append(window)
        y_list.append(y_t)
        used_dt.append(dt)
    X = np.stack(X_list) if X_list else np.empty((0, TIME_WINDOW, num_nodes, num_features), dtype=np.float32)
    y = np.stack(y_list) if y_list else np.empty((0, num_nodes), dtype=np.float32)
    return X, y, used_dt


X_train, y_train, dt_train = build_train_samples(train_valid_timesteps)

print(f"X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"Jumlah timestep TEST yang akan diprediksi (autoregresif): {len(test_valid_timesteps)}")
```

### Split Train/Test Sesuai Periode & Tensor Final

## TRAINING PROCESS

```python
import os
import time
import re
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

# ======================================================================
# 0) Device & multi-GPU setup (2x T4 di Kaggle)
# ======================================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
n_gpus = torch.cuda.device_count()
print(f"Device: {device} | Jumlah GPU terdeteksi: {n_gpus}")
for i in range(n_gpus):
    print(f"  GPU {i}: {torch.cuda.get_device_name(i)}")

USE_MULTI_GPU = n_gpus >= 2

# ======================================================================
# 1) Load dataset hasil pipeline sebelumnya
# ======================================================================
DATA_PATH = "/kaggle/working/gnn_train_test_dataset.npz"  # sesuaikan jika beda
data = np.load(DATA_PATH, allow_pickle=True)

X_train_all = data["X_train"]        # (N_train, time_window, num_nodes, num_features)
y_train_all = data["y_train"]        # (N_train, num_nodes)
edge_index_np = data["edge_index"]   # (2, num_edges)
edge_weight_np = data["edge_weight"] # (num_edges,)
node_order = list(data["node_order"])
feature_cols = list(data["feature_cols"])
dt_train_all = pd.to_datetime(data["dt_train"])
panel_arr_all = data["panel_arr"].copy()
obs_datetimes_all = pd.to_datetime(data["obs_datetimes"])

num_nodes = X_train_all.shape[2]
num_features = X_train_all.shape[3]
time_window = X_train_all.shape[1]

print(f"X_train_all: {X_train_all.shape}, y_train_all: {y_train_all.shape}")
print(f"num_nodes={num_nodes}, num_features={num_features}, time_window={time_window}")

if X_train_all.shape[0] == 0:
    raise ValueError("X_train kosong! Cek ulang tahap pembuatan dataset sebelum training.")

# ======================================================================
# 2) Split waktu dulu, lalu fit semua scaler HANYA pada bagian train.
#    Tinggi air dinormalisasi per-pos karena datum antarlokasi sangat beda.
# ======================================================================
n_train_total = X_train_all.shape[0]
val_ratio = 0.1
n_val = max(1, int(n_train_total * val_ratio))
n_tr = n_train_total - n_val

wl_t_pos = feature_cols.index("wl_t")
wl_value_positions = [i for i, c in enumerate(feature_cols) if (
    c == "wl_t" or c.startswith("wl_t-") or
    (c.startswith("wl_roll") and not c.endswith("_std"))
)]
wl_std_positions = [i for i, c in enumerate(feature_cols) if c.startswith("wl_roll") and c.endswith("_std")]

# Robust center/scale per pos. IQR lebih tahan terhadap spike sensor.
wl_train_values = X_train_all[:n_tr, :, :, wl_t_pos].reshape(-1, num_nodes)
station_center = np.nanmedian(wl_train_values, axis=0).astype(np.float32)
q25 = np.nanquantile(wl_train_values, 0.25, axis=0)
q75 = np.nanquantile(wl_train_values, 0.75, axis=0)
station_scale = (q75 - q25).astype(np.float32)
station_std = np.nanstd(wl_train_values, axis=0).astype(np.float32)
station_scale = np.where(station_scale > 1e-3, station_scale, station_std)
station_scale = np.where(station_scale > 1e-3, station_scale, 1.0).astype(np.float32)

# Batas fisik empiris longgar untuk mencegah recursive drift puluhan meter.
station_q01 = np.nanquantile(wl_train_values, 0.01, axis=0).astype(np.float32)
station_q99 = np.nanquantile(wl_train_values, 0.99, axis=0).astype(np.float32)
station_low = station_q01 - 0.50 * station_scale
station_high = station_q99 + 0.50 * station_scale

def station_normalize_water_features(X):
    X = X.copy()
    center = station_center.reshape(1, 1, num_nodes)
    scale = station_scale.reshape(1, 1, num_nodes)
    for p in wl_value_positions:
        X[..., p] = (X[..., p] - center) / scale
    for p in wl_std_positions:
        X[..., p] = X[..., p] / scale
    return X

X_station_n = station_normalize_water_features(X_train_all)
X_fit_flat = X_station_n[:n_tr].reshape(-1, num_features)
feat_mean = X_fit_flat.mean(axis=0).astype(np.float32)
feat_std = X_fit_flat.std(axis=0).astype(np.float32)
feat_std[feat_std < 1e-6] = 1.0

def normalize_X(X):
    return (station_normalize_water_features(X) - feat_mean) / feat_std

def normalize_y(y):
    return (y - station_center.reshape(1, -1)) / station_scale.reshape(1, -1)

def denormalize_y(y):
    return y * station_scale.reshape(1, -1) + station_center.reshape(1, -1)

X_train_all_n = ((X_station_n - feat_mean) / feat_std).astype(np.float32)
y_train_all_n = normalize_y(y_train_all).astype(np.float32)

# ======================================================================
# 3) Split train/validation berbasis waktu.
# ======================================================================
X_tr, y_tr = X_train_all_n[:n_tr], y_train_all_n[:n_tr]
X_val, y_val = X_train_all_n[n_tr:], y_train_all_n[n_tr:]

print(f"Train: {X_tr.shape[0]} sample | Val: {X_val.shape[0]} sample")

# ======================================================================
# 4) Dataset & DataLoader
# ======================================================================
class GraphTimeSeriesDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.from_numpy(X)  # (N, T, num_nodes, num_features)
        self.y = torch.from_numpy(y)  # (N, num_nodes)

    def __len__(self):
        return self.X.shape[0]

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

BATCH_SIZE = 64
train_ds = GraphTimeSeriesDataset(X_tr, y_tr)
val_ds = GraphTimeSeriesDataset(X_val, y_val)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, drop_last=False)
val_loader = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False)

# ======================================================================
# 5) Bangun adjacency ternormalisasi (dense, num_nodes kecil ~30 -> aman)
# ======================================================================
def build_normalized_adjacency(edge_index, edge_weight, num_nodes):
    A = np.zeros((num_nodes, num_nodes), dtype=np.float32)
    for k in range(edge_index.shape[1]):
        i, j = edge_index[0, k], edge_index[1, k]
        w = edge_weight[k] if edge_weight is not None else 1.0
        w = 1.0 / (w + 1e-3)  # edge lebih pendek -> bobot lebih besar
        A[i, j] = max(A[i, j], w)
    A = A + np.eye(num_nodes, dtype=np.float32)  # self-loop
    deg = A.sum(axis=1)
    deg_inv_sqrt = np.power(deg, -0.5, where=deg > 0)
    deg_inv_sqrt[deg == 0] = 0.0
    D_inv_sqrt = np.diag(deg_inv_sqrt)
    A_norm = D_inv_sqrt @ A @ D_inv_sqrt
    return A_norm.astype(np.float32)

A_norm_np = build_normalized_adjacency(edge_index_np, edge_weight_np, num_nodes)

# ======================================================================
# 6) Model: GCN per-timestep (shared weight) + GRU temporal + head
# ======================================================================
class GCNLayer(nn.Module):
    def __init__(self, in_dim, out_dim):
        super().__init__()
        self.linear = nn.Linear(in_dim, out_dim)

    def forward(self, x, A_norm):
        # x: (B, num_nodes, in_dim) ; A_norm: (num_nodes, num_nodes)
        x = torch.einsum("ij,bjf->bif", A_norm, x)
        return self.linear(x)


class TemporalGNN(nn.Module):
    def __init__(self, num_nodes, num_features, A_norm,
                 gcn_hidden=64, gru_hidden=64, dropout=0.2):
        super().__init__()
        self.register_buffer("A_norm", torch.from_numpy(A_norm))
        self.num_nodes = num_nodes

        self.gcn1 = GCNLayer(num_features, gcn_hidden)
        self.local_skip = nn.Linear(num_features, gcn_hidden)
        self.gcn2 = GCNLayer(gcn_hidden, gcn_hidden)
        self.act = nn.ReLU()
        self.drop = nn.Dropout(dropout)

        self.gru = nn.GRU(input_size=gcn_hidden, hidden_size=gru_hidden, batch_first=True)
        self.head = nn.Sequential(
            nn.Linear(gru_hidden, gru_hidden // 2),
            nn.ReLU(),
            nn.Linear(gru_hidden // 2, 1),
        )

    def forward(self, X):
        # X: (B, T, num_nodes, num_features)
        B, T, N, F = X.shape
        gcn_out = []
        for t in range(T):
            # Residual lokal menjaga sinyal pos sendiri agar tidak larut oleh tetangga.
            h = self.gcn1(X[:, t], self.A_norm) + self.local_skip(X[:, t])
            h = self.act(h)
            h = self.drop(h)
            h = self.gcn2(h, self.A_norm)
            h = self.act(h)
            gcn_out.append(h)  # (B, N, gcn_hidden)
        h_seq = torch.stack(gcn_out, dim=1)              # (B, T, N, gcn_hidden)

        # reshape supaya GRU jalan per-node: (B*N, T, gcn_hidden)
        h_seq = h_seq.permute(0, 2, 1, 3).reshape(B * N, T, -1)
        _, h_n = self.gru(h_seq)                          # h_n: (1, B*N, gru_hidden)
        h_last = h_n.squeeze(0)                            # (B*N, gru_hidden)

        out = self.head(h_last).view(B, N)                 # (B, num_nodes)
        return out


model = TemporalGNN(num_nodes, num_features, A_norm_np,
                     gcn_hidden=64, gru_hidden=64, dropout=0.2)

if USE_MULTI_GPU:
    print(f"Memakai nn.DataParallel di {n_gpus} GPU: {list(range(n_gpus))}")
    model = nn.DataParallel(model, device_ids=list(range(n_gpus)))

model = model.to(device)

# ======================================================================
# 7) Training loop
# ======================================================================
EPOCHS = 100
LR = 1e-3
PATIENCE = 15
CKPT_PATH = "/kaggle/working/best_temporal_gnn.pt"

# Huber lebih tahan terhadap spike/outlier sensor daripada MSE murni.
criterion = nn.SmoothL1Loss(beta=0.5)
optimizer = torch.optim.Adam(model.parameters(), lr=LR, weight_decay=1e-5)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode="min", factor=0.5, patience=5)

def run_epoch(loader, train=True):
    model.train() if train else model.eval()
    total_loss, n_batches = 0.0, 0
    with torch.set_grad_enabled(train):
        for X_batch, y_batch in loader:
            X_batch = X_batch.to(device, non_blocking=True)
            y_batch = y_batch.to(device, non_blocking=True)

            if train:
                optimizer.zero_grad()
            pred = model(X_batch)
            loss = criterion(pred, y_batch)

            if train:
                loss.backward()
                optimizer.step()

            total_loss += loss.item()
            n_batches += 1
    return total_loss / max(n_batches, 1)

# Validation rollout: meniru test dengan memasukkan prediksi kembali sebagai histori.
lag_positions = [feature_cols.index(f"wl_t-{k}") for k in range(1, 12)]
roll_window_sizes = sorted({
    int(m.group(1)) for c in feature_cols
    if (m := re.match(r"wl_roll(\d+)_(mean|std|min|max)$", c))
})
roll_specs = {w: {
    stat: feature_cols.index(f"wl_roll{w}_{stat}")
    for stat in ("mean", "std", "min", "max")
} for w in roll_window_sizes}
datetime_to_pos_all = {dt: i for i, dt in enumerate(obs_datetimes_all)}
actual_wl_all = panel_arr_all[:, :, wl_t_pos].copy()

def rebuild_water_window(window, start_pos, wl_series):
    """Rebuild fitur air dari histori aktual/prediksi secara kausal."""
    for offset, k in enumerate(range(start_pos, start_pos + window.shape[0])):
        window[offset, :, wl_t_pos] = wl_series[k]
        for lag_idx, lag_col_pos in enumerate(lag_positions, start=1):
            if k - lag_idx >= 0:
                window[offset, :, lag_col_pos] = wl_series[k - lag_idx]
        for w, cols in roll_specs.items():
            if k - w >= 0:
                vals = wl_series[k - w:k]
                window[offset, :, cols["mean"]] = vals.mean(axis=0)
                window[offset, :, cols["std"]] = vals.std(axis=0, ddof=1)
                window[offset, :, cols["min"]] = vals.min(axis=0)
                window[offset, :, cols["max"]] = vals.max(axis=0)
    return window

def run_rollout_validation():
    base = model.module if isinstance(model, nn.DataParallel) else model
    base.eval()
    rollout_wl = actual_wl_all.copy()
    val_input_datetimes = pd.DatetimeIndex(dt_train_all[n_tr:])
    first_input_pos = datetime_to_pos_all[val_input_datetimes[0]]
    rollout_wl[first_input_pos + 1:] = np.nan
    pred_list, true_list = [], []

    with torch.no_grad():
        for dt_input in val_input_datetimes:
            input_pos = datetime_to_pos_all[dt_input]
            target_pos = input_pos + 1
            if target_pos >= len(obs_datetimes_all):
                continue
            # Jika valid timestep terputus akibat missing data, mulai segmen baru
            # dari observasi aktual terakhir; tidak menyelundupkan nilai dalam segmen.
            if np.isnan(rollout_wl[input_pos]).any():
                rollout_wl[:input_pos + 1] = actual_wl_all[:input_pos + 1]
            start = input_pos - (time_window - 1)
            if start < 0:
                continue
            window = panel_arr_all[start:input_pos + 1].copy()
            window = rebuild_water_window(window, start, rollout_wl)
            if np.isnan(window).any() or np.isnan(actual_wl_all[target_pos]).any():
                continue
            X_batch = torch.from_numpy(normalize_X(window[None]).astype(np.float32)).to(device)
            pred_n = base(X_batch).cpu().numpy()[0]
            pred = pred_n * station_scale + station_center
            pred = np.clip(pred, station_low, station_high)
            rollout_wl[target_pos] = pred
            pred_list.append(pred)
            true_list.append(actual_wl_all[target_pos])

    if not pred_list:
        return float("inf")
    pred_arr = np.stack(pred_list)
    true_arr = np.stack(true_list)
    return float(np.sqrt(np.mean(((pred_arr - true_arr) / station_scale) ** 2)))

best_val_loss = float("inf")
patience_counter = 0

print("\n=== Mulai training ===")
for epoch in range(1, EPOCHS + 1):
    t0 = time.time()
    train_loss = run_epoch(train_loader, train=True)
    teacher_val_loss = run_epoch(val_loader, train=False)
    rollout_val_loss = run_rollout_validation()
    scheduler.step(rollout_val_loss)
    dt = time.time() - t0

    # Checkpoint dipilih dari rollout, bukan teacher-forced one-step loss.
    improved = rollout_val_loss < best_val_loss
    if improved:
        best_val_loss = rollout_val_loss
        patience_counter = 0
        state_dict = model.module.state_dict() if USE_MULTI_GPU else model.state_dict()
        torch.save({
            "model_state_dict": state_dict,
            "feat_mean": feat_mean, "feat_std": feat_std,
            "station_center": station_center, "station_scale": station_scale,
            "station_low": station_low, "station_high": station_high,
            "wl_value_positions": np.array(wl_value_positions),
            "wl_std_positions": np.array(wl_std_positions),
            "A_norm": A_norm_np,
            "num_nodes": num_nodes, "num_features": num_features,
            "time_window": time_window,
        }, CKPT_PATH)
    else:
        patience_counter += 1

    print(f"Epoch {epoch:3d}/{EPOCHS} | train_loss={train_loss:.5f} | "
          f"teacher_val={teacher_val_loss:.5f} | rollout_val={rollout_val_loss:.5f} | "
          f"lr={optimizer.param_groups[0]['lr']:.2e} | "
          f"{dt:.1f}s" + ("  <- best" if improved else ""))

    if patience_counter >= PATIENCE:
        print(f"Early stopping di epoch {epoch} (val_loss tidak membaik {PATIENCE}x).")
        break

print(f"\nBest rollout_val (normalized RMSE): {best_val_loss:.5f} | checkpoint: {CKPT_PATH}")

# ======================================================================
# 8) Evaluasi cepat di validation set (skala asli / denormalized)
#    (evaluasi TEST tidak dilakukan di sini — test butuh inference
#     autoregresif terpisah, lihat cell inference)
# ======================================================================
ckpt = torch.load(CKPT_PATH, map_location=device, weights_only=False)
base_model = TemporalGNN(num_nodes, num_features, ckpt["A_norm"],
                          gcn_hidden=64, gru_hidden=64, dropout=0.0)
base_model.load_state_dict(ckpt["model_state_dict"])
base_model = base_model.to(device)
base_model.eval()

all_preds, all_targets = [], []
with torch.no_grad():
    for X_batch, y_batch in val_loader:
        X_batch = X_batch.to(device)
        pred = base_model(X_batch).cpu().numpy()
        all_preds.append(pred)
        all_targets.append(y_batch.numpy())

y_pred_n = np.concatenate(all_preds, axis=0)
y_true_n = np.concatenate(all_targets, axis=0)

y_pred = denormalize_y(y_pred_n)
y_true = denormalize_y(y_true_n)

rmse = np.sqrt(np.mean((y_pred - y_true) ** 2))
mae = np.mean(np.abs(y_pred - y_true))

print("\n=== Evaluasi Validation Set (skala asli tma_mdpl) ===")
print(f"RMSE: {rmse:.4f}")
print(f"MAE : {mae:.4f}")

per_node_rmse = np.sqrt(np.mean((y_pred - y_true) ** 2, axis=0))
for i, name in enumerate(node_order):
    print(f"  {name:35s} RMSE={per_node_rmse[i]:.4f}")
```

## INFERENCE

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import re

# ======================================================================
# 0) Device
# ======================================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# ======================================================================
# 1) Definisi ulang model (harus identik dengan yang dipakai saat training)
# ======================================================================
class GCNLayer(nn.Module):
    def __init__(self, in_dim, out_dim):
        super().__init__()
        self.linear = nn.Linear(in_dim, out_dim)

    def forward(self, x, A_norm):
        x = torch.einsum("ij,bjf->bif", A_norm, x)
        return self.linear(x)


class TemporalGNN(nn.Module):
    def __init__(self, num_nodes, num_features, A_norm,
                 gcn_hidden=64, gru_hidden=64, dropout=0.2):
        super().__init__()
        self.register_buffer("A_norm", torch.from_numpy(A_norm))
        self.num_nodes = num_nodes
        self.gcn1 = GCNLayer(num_features, gcn_hidden)
        self.local_skip = nn.Linear(num_features, gcn_hidden)
        self.gcn2 = GCNLayer(gcn_hidden, gcn_hidden)
        self.act = nn.ReLU()
        self.drop = nn.Dropout(dropout)
        self.gru = nn.GRU(input_size=gcn_hidden, hidden_size=gru_hidden, batch_first=True)
        self.head = nn.Sequential(
            nn.Linear(gru_hidden, gru_hidden // 2),
            nn.ReLU(),
            nn.Linear(gru_hidden // 2, 1),
        )

    def forward(self, X):
        B, T, N, F = X.shape
        gcn_out = []
        for t in range(T):
            h = self.gcn1(X[:, t], self.A_norm) + self.local_skip(X[:, t])
            h = self.act(h)
            h = self.drop(h)
            h = self.gcn2(h, self.A_norm)
            h = self.act(h)
            gcn_out.append(h)
        h_seq = torch.stack(gcn_out, dim=1)
        h_seq = h_seq.permute(0, 2, 1, 3).reshape(B * N, T, -1)
        _, h_n = self.gru(h_seq)
        h_last = h_n.squeeze(0)
        out = self.head(h_last).view(B, N)
        return out

# ======================================================================
# 2) Load checkpoint & dataset (panel_arr = fitur mentah seluruh horizon)
# ======================================================================
CKPT_PATH = "/kaggle/working/best_temporal_gnn.pt"
DATA_PATH = "/kaggle/working/gnn_train_test_dataset.npz"
SAMPLE_SUB_PATH = "/kaggle/input/competitions/sebelas-maret-statistics-data-science-2026/sample_submission.csv"
OUTPUT_PATH = "/kaggle/working/submission.csv"

N_LAGS = 11          # HARUS sama dengan saat membangun dataset
TIME_WINDOW = 12     # HARUS sama dengan saat training
TRAIN_END = "2025-09-18 18:00:00"   # HARUS sama dengan saat membangun dataset

ckpt = torch.load(CKPT_PATH, map_location=device, weights_only=False)
feat_mean = ckpt["feat_mean"]
feat_std = ckpt["feat_std"]
station_center = ckpt["station_center"]
station_scale = ckpt["station_scale"]
station_low = ckpt["station_low"]
station_high = ckpt["station_high"]
wl_value_positions = list(ckpt["wl_value_positions"])
wl_std_positions = list(ckpt["wl_std_positions"])
A_norm_np = ckpt["A_norm"]
num_nodes = ckpt["num_nodes"]
num_features = ckpt["num_features"]

data = np.load(DATA_PATH, allow_pickle=True)
panel_arr = data["panel_arr"].copy()                       # (T_total, num_nodes, num_features)
obs_datetimes_all = pd.to_datetime(data["obs_datetimes"])
feature_cols = list(data["feature_cols"])
node_order = list(data["node_order"])

wl_t_pos = feature_cols.index("wl_t")
lag_positions = [feature_cols.index(f"wl_t-{k}") for k in range(1, N_LAGS + 1)]
datetime_to_pos = {dt: i for i, dt in enumerate(obs_datetimes_all)}

def normalize_inference_window(window):
    X = window.copy()
    center = station_center.reshape(1, 1, num_nodes)
    scale = station_scale.reshape(1, 1, num_nodes)
    for p in wl_value_positions:
        X[..., p] = (X[..., p] - center) / scale
    for p in wl_std_positions:
        X[..., p] = X[..., p] / scale
    return (X - feat_mean) / feat_std

infer_model = TemporalGNN(num_nodes, num_features, A_norm_np,
                           gcn_hidden=64, gru_hidden=64, dropout=0.0)
infer_model.load_state_dict(ckpt["model_state_dict"])
infer_model = infer_model.to(device).eval()

# ======================================================================
# 3) Tentukan target datetime yang perlu diprediksi (langsung dari submission)
# ======================================================================
sample_sub = pd.read_csv(SAMPLE_SUB_PATH)
sample_sub["datetime"] = pd.to_datetime(sample_sub["id"].str[:19])
sample_sub["nama_pos"] = sample_sub["id"].str[22:]

target_datetimes = pd.DatetimeIndex(sorted(sample_sub["datetime"].unique()))
print(f"Jumlah timestep TARGET (perlu diprediksi): {len(target_datetimes)}")
print(f"Rentang target: {target_datetimes.min()} s/d {target_datetimes.max()}")

# ======================================================================
# 4) wl_series: gabungan wl aktual (train) + wl hasil prediksi (test),
#    diupdate maju seiring waktu (causal, autoregresif)
# ======================================================================
wl_series = panel_arr[:, :, wl_t_pos].copy()   # (T_total, num_nodes) - NaN di test sebelum diprediksi

# --- Tutup gap histori hanya secara kausal: dilarang memakai bfill. ---
last_train_pos = datetime_to_pos[pd.Timestamp(TRAIN_END)]
n_filled = 0
for node_i in range(num_nodes):
    col = wl_series[:last_train_pos + 1, node_i]
    if np.isnan(col).any():
        n_filled += np.isnan(col).sum()
        # Jangan gunakan observasi masa depan untuk mengisi histori.
        s = pd.Series(col).ffill()
        wl_series[:last_train_pos + 1, node_i] = s.values
print(f"Total sel wl_t (periode train) yang di-forward-fill secara kausal: {n_filled}")

# --- siapkan posisi kolom rolling, window size diambil otomatis dari feature_cols ---
roll_window_sizes = sorted(set(
    int(m.group(1)) for c in feature_cols
    if (m := re.match(r"wl_roll(\d+)_(mean|std|min|max)$", c))
))
roll_specs = {}
for w in roll_window_sizes:
    roll_specs[w] = {
        "mean": feature_cols.index(f"wl_roll{w}_mean"),
        "std":  feature_cols.index(f"wl_roll{w}_std"),
        "min":  feature_cols.index(f"wl_roll{w}_min"),
        "max":  feature_cols.index(f"wl_roll{w}_max"),
    }
print(f"Rolling window sizes terdeteksi: {roll_window_sizes}")

# Audit eksplisit sambungan train -> test. Hanya histori yang diperlukan
# oleh window test pertama yang harus lengkap; NaN awal deret tidak relevan.
first_target_pos = datetime_to_pos[target_datetimes.min()]
first_window_start = first_target_pos - TIME_WINDOW
history_start = max(0, first_window_start - max(N_LAGS, max(roll_window_sizes)))
required_history = wl_series[history_start:first_target_pos]
assert not np.isnan(required_history).any(), (
    "Histori tinggi air di batas train -> test masih kosong; "
    "periksa data train, jangan mengisi dengan informasi masa depan."
)
print(
    "Sambungan train -> test tervalidasi: "
    f"window pertama {obs_datetimes_all[first_window_start]} s/d "
    f"{obs_datetimes_all[first_target_pos - 1]} memakai histori train."
)

predictions = {}   # {target_datetime: array (num_nodes,)} skala asli tma_mdpl
skipped = []

with torch.no_grad():
    for dt_target in target_datetimes:
        if dt_target not in datetime_to_pos:
            skipped.append(dt_target)
            continue

        t_target_pos = datetime_to_pos[dt_target]
        t_end_pos = t_target_pos - 1          # window berakhir SATU LANGKAH sebelum target
        start = t_end_pos - (TIME_WINDOW - 1)

        if t_end_pos < 0 or start < 0:
            skipped.append(dt_target)
            continue

        window = panel_arr[start:t_end_pos + 1].copy()   # (TIME_WINDOW, num_nodes, num_features)

        for offset, k in enumerate(range(start, t_end_pos + 1)):
            window[offset, :, wl_t_pos] = wl_series[k]
            for lag_idx, lag_col_pos in enumerate(lag_positions, start=1):
                src_k = k - lag_idx
                if src_k >= 0:
                    window[offset, :, lag_col_pos] = wl_series[src_k]

            for w, cols in roll_specs.items():
                roll_start = k - w
                if roll_start >= 0:
                    roll_vals = wl_series[roll_start:k]          # (w, num_nodes)
                    window[offset, :, cols["mean"]] = roll_vals.mean(axis=0)
                    window[offset, :, cols["std"]]  = roll_vals.std(axis=0, ddof=1)
                    window[offset, :, cols["min"]]  = roll_vals.min(axis=0)
                    window[offset, :, cols["max"]]  = roll_vals.max(axis=0)

        if np.isnan(window).any():
            skipped.append(dt_target)
            continue

        window_n = normalize_inference_window(window[None])[0]
        X_batch = torch.from_numpy(window_n[None].astype(np.float32)).to(device)

        pred_n = infer_model(X_batch).cpu().numpy()[0]
        pred_raw = pred_n * station_scale + station_center
        # Guardrail long-horizon: mencegah drift di luar rentang historis wajar.
        pred = np.clip(pred_raw, station_low, station_high)

        predictions[dt_target] = pred
        wl_series[t_target_pos] = pred          # simpan sbg "wl aktual" utk target berikutnya

print(f"Berhasil diprediksi: {len(predictions)} / {len(target_datetimes)} timestep")
if skipped:
    print(f"Dilewati (fitur masih NaN / di luar range): {len(skipped)} -> {skipped[:5]} ...")

# ======================================================================
# 4b) Rangkai predictions -> pred_df   <-- INI YANG HILANG DI KODE ANDA
# ======================================================================
pred_df = pd.DataFrame(predictions).T
pred_df.columns = node_order
pred_df.index.name = "datetime"
pred_df = pred_df.sort_index()
print(pred_df.shape)
pred_df.head()

# ======================================================================
# 5) Susun ke format submission
#    id = "YYYY-MM-DD HH:MM:SS - nama_pos"
# ======================================================================
pred_long = pred_df.reset_index().melt(id_vars="datetime", var_name="nama_pos", value_name="tma_mdpl_pred")

submission = sample_sub.merge(pred_long, on=["datetime", "nama_pos"], how="left")

n_missing = submission["tma_mdpl_pred"].isna().sum()
if n_missing > 0:
    print(f"Peringatan: {n_missing} baris tidak ada prediksi, akan diisi fallback.")
    submission = submission.sort_values(["nama_pos", "datetime"])
    submission["tma_mdpl_pred"] = submission.groupby("nama_pos")["tma_mdpl_pred"].ffill()
    fallback_mean = submission.groupby("nama_pos")["tma_mdpl_pred"].transform("mean")
    submission["tma_mdpl_pred"] = submission["tma_mdpl_pred"].fillna(fallback_mean)

submission_final = sample_sub[["id"]].merge(
    submission[["id", "tma_mdpl_pred"]], on="id", how="left"
)
submission_final = submission_final.rename(columns={"tma_mdpl_pred": "tma_mdpl"})

assert submission_final.shape[0] == sample_sub.shape[0], "Jumlah baris tidak cocok dengan sample_submission!"
assert submission_final["tma_mdpl"].isna().sum() == 0, "Masih ada NaN di kolom prediksi!"

submission_final.to_csv(OUTPUT_PATH, index=False)
print(f"Submission disimpan ke: {OUTPUT_PATH}")
submission_final.head()
```
